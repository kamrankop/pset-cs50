```python
# c) i)

# Pooled OLS Modela and generate the standard error using clustering model

PooledOLSclustered = smf.ols("dem_ind ~ log_gdppc", data=demdf)

results = PooledOLSclustered.fit(cov_type='cluster',
                                 cov_kwds={'groups': demdf[['country', 'year', 'dem_ind', 'log_gdppc']].dropna()['country']})

results.summary()

# ANSWER: our co-efficient of interest is 0.2357, and it is statistically significant.
```

```python
# c) ii)

0.2357*0.2
```

```python
# for 95% confidence interval
lb = 0.2*(0.2357-1.96*0.012)
lb
```

```python
ub = 0.2*(0.2357+1.96*0.012)
ub
```

```python
# c) ii) ANSWER: Our predicted change (coefficient x 0.20) is 0.04714, and is statistically significant.
#                Our Confidence Interval is 0.0424, statistically significant coefficient.
```

```python
# c) iii) Unclustured

PooledOLS_Unclustered = smf.ols("dem_ind ~ log_gdppc", data = demdf)

results2 = PooledOLS_Unclustered.fit(cov_type='HC1')
results2.summary()

```

```python
# c) iii) ANSWER: log_gdppc = 0.2357, same as before. Std_error = 0.0007

# We are going to reject the null hypothesis
```

```python
# D) i)

# ANSWER Culture is very persistent and seeps into customs and institutions over a long period of time.
# This could be the cause of Omitted Variable Bias in (c)
```

```python
# d) ii)

Panel_data = demdf.set_index(['country', 'year'])

print('Panel_data')
Panel_data
```

```python
PanelFE = PanelOLS(Panel_data.dem_ind, Panel_data.log_gdppc, entity_effects = True)

PanelFE_fitted = PanelFE.fit(cov_type = 'clustered', cluster_entity = True, cluster_time = False)

print(PanelFE_fitted)
```

```python
# ANSWER here our coefficient of interest is no longer 0.25, but is now reduced to 0.0837. The T-stat of 2.665 shows it is still significant.
```

```python
# d) iii) exclide the results for azerbaijan and compare results

data_noAz = Panel_data.drop(labels = "Azerbaijan", axis="index", level = "country")

data_noAz
```

```python
# d) iii) PanelFE_noAz = PanelOLS(data_noAz.dem_ind, data_noAz.log_gdppc, entity_effects = True)

print(PanelFE_noAz.fit(cov_type ='clustered', cluster_entity = True, cluster_time = False))
```

```python
# d) iv) What variable could cause Omitted Variable Bias?

#ANSWER: The demand for democracy coincides with the changes in economic conditions,
#        this is correlated with the log_gdppc and could cause the omitted variable bias.
```

```python
# d) v)  Estimate for time fixed effects

twoway = PanelOLS(Panel_data.dem_ind, Panel_data.log_gdppc, entity_effects = True, time_effects = True)

print(twoway.fit(cov_type='clustered', cluster_entity = True, cluster_time = False))
```

```python
# ANSWER d) v)

# The standard error generated, 0.046, is just slightly larger than that of the authors and that
# generated by R, 0.042. the coefficient of interest is no longer significant when we include the time effects (t-stat is well below 2)


# there may have been omitted variable bias as our variable of interest coeff is changing
```
