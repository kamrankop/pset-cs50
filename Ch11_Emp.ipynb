```python
Ch11data = pd.read_excel(r'/content/smoking_kk.xlsx', sheet_name = 'Smoking')
print(Ch11data)
```

```python
df=pd.DataFrame(Ch11data)
df
```

```python
# a)

# probabilities will be estimated below by requesting the means

Ch11data[["smoker"]].mean()
```

```python
means = Ch11data.groupby('smkban')['smoker'].mean()
means
```

```python
# b)

0.289595-0.212037
```

```python
lin_prob = smf.ols("smoker ~ smkban", data=df)
results = lin_prob.fit(cov_type='HC1')
results.summary()
```

```python
# ANSWER b) The beta1 smkban coeff, and the z-stat(-8.664) show that this is statistically  significant
```

```python
# c)

df["age2"] = df["age"]**2
print(df)
```

```python
lin_prob_multi = smf.ols("smoker ~ smkban+female+age+age2+hsdrop+hsgrad+colsome+colgrad+black+hispanic",
                         data=df)
lin_prob_results = lin_prob_multi.fit(cov_type='HC1')
lin_prob_results.summary()
```

```python
#ANSWER c) the co-eff of interest smkban (-0.0472) is smaller than above results.
#           smkban may be correlated with the education/race/gender indicators or with age.
```

```python
#ANSWER d) our z-statistic is -5.269, so the coefficient is statistically significant at the 1% level.
```

```python
# e) This is a joint hypothesis test. For this we run our f-test

hypotheses = '(hsdrop = 0), (hsgrad = 0), (colsome = 0), (colgrad = 0)'
f_test = lin_prob_results.f_test(hypotheses)
print(f_test)
```

```python
#ANSWER e) F-statistic is 0.00, i.e our coefficients are statistically significant.
```

```python
# f) estimate the probit model


Probit = smf.probit("smoker ~ smkban+female+age+age2+hsdrop+hsgrad+colsome+colgrad+black+hispanic", data=df)
Probit = Probit.fit(cov_type='HC1')
Probit.summary()
```

```python
#ANSWER f) The t-statistic for the coefficient on smkban is -5.47. As this value is similar to the value of our linear prob. model.

#           Similarly, we conclude the F-statistic is significant at the 1% level
```

```python
# g) Estimate the Logit model.

Logit = smf.logit("smoker ~ smkban+female+age+age2+hsdrop+hsgrad+colsome+colgrad+black+hispanic", data=df)
Logit = Logit.fit(cov_type='HC1')
Logit.summary()
```

```python
#ANSWER g) Again, this is similar to our above probit and linear prob. model, where the t-statistic for the coefficient on smkban
#          is -5.31 is similar. We conclude that the f-statistic is significant at 1% level.
```
