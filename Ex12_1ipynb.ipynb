```python
## Question Part(F): Estimate the IV regression of 'weeksworked' on 'morekids', using 'samesex' as an instrument. How large is the fertility effect on labor supply?
```

```python
## ANSWER Part(F):
import statsmodels.api as sm

dependent = Ch12data.weeksm1
controls = sm.add_constant(Ch12data[[]])
endog = Ch12data[['morekids']]
instruments = Ch12data[['samesex']]

# call results:

PartF = IV2SLS(dependent, controls, endog, instruments)
PartFfitted = PartF.fit(cov_type = 'robust')
print(PartFfitted)
```

```python
## ANSWER Part(F):

## We get a co-efficient of interest on our variable = -6.0332.

## This means 6 hours less per work if they have more than two kids
```

```python
## Question Part(G): Do the results change when you include the variables 'agem1', 'black', 'hispan', 'othrace' in the labor supply regression?
```

```python
## ANSWER Part(G):

dependent = Ch12data.weeksm1
controls = sm.add_constant(Ch12data[['agem1', 'black', 'hispan', 'othrace']])
endog = Ch12data[['morekids']]
instruments = Ch12data[['samesex']]

# call results:

PartG = IV2SLS(dependent, controls, endog, instruments)
PartGfitted = PartG.fit(cov_type = 'robust')
print(PartGfitted)
```

```python
## ANSWER Part(G):

## We get a co-efficient of interest on our variable 'morekids'= -5.78, with a t-stat of -1.6

## As the result; the change in the coefficient 'morekids', is less than one std dev, this suggests a modest omitted variable bias.

## This shows us that the results after includes race-related variables do NOT change by a large margin.
## This is because 'samesex' is unrelated to the ageml, black, hispan, othrace variables.
```
